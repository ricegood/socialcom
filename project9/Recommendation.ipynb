```python
'''HW1'''
'''피어슨 상관계수'''
import math

userdict = {}
userdict['user1'] = {'item1':4, 'item2':3, 'item3':4, 'item6':1}
userdict['user2'] = {'item1':4, 'item2':1, 'item3':2, 'item4':2, 'item6':1}
userdict['user3'] = {'item1':4, 'item4':3, 'item5':1}
userdict['user4'] = {'item1':1, 'item2':4, 'item3':2, 'item5':4}
userdict['user5'] = {'item1':1, 'item3':3, 'item4':1, 'item5':3, 'item6':2}

useravg = {}
for user in userdict:
    sum = 0
    for item in userdict[user]:
        sum += userdict[user][item]
    useravg[user] = sum / len(userdict[user])
    
x = 'user1'
xavg = useravg[x]
for y in userdict:
    yavg = useravg[y]
    term1 = 0
    term2 = 0
    term3 = 0
    for item in userdict[x]:
        if item in userdict[y]:
            rxs = userdict[x][item]
            rys = userdict[y][item]
            term1 += (rxs-xavg)*(rys-yavg)
            term2 += (rxs-xavg)**2
            term3 += (rys-yavg)**2
    term4 = (term2**(1/2))*(term3**(1/2))
    if term4 == 0:
        print('sim(', x, ',', y,') = INF')
    else:
        sim = term1/term4
        print('sim(', x, ',', y,') = ', sim)
```

```python
'''EXAMPLE'''
from scipy import spatial

userdict = {}
userdict['user1'] = {'item1':1, 'item3':2, 'item6':1}
userdict['user2'] = {'item3':4, 'item4':2}
userdict['user3'] = {'item1':3, 'item2':5, 'item4':4, 'item5':4, 'item6':3}
userdict['user4'] = {'item2':4, 'item3':1, 'item5':3}
userdict['user5'] = {'item3':2, 'item4':5, 'item5':4, 'item6':3}
userdict['user6'] = {'item1':5, 'item5':2}
userdict['user7'] = {'item2':4, 'item3':3}
userdict['user8'] = {'item4':4, 'item6':2}
userdict['user9'] = {'item1':5, 'item3':4}
userdict['user10'] = {'item2':2, 'item3':3}
userdict['user11'] = {'item1':4, 'item2':1, 'item3':5, 'item4':2, 'item5':2, 'item6':4}
userdict['user12'] = {'item2':3, 'item5':5}

itemdict = {}
for i in range(1, 7):
    item = 'item'+str(i)
    itemdict[item] = {}
    for user in userdict:
        if item in userdict[user]:
            itemdict[item][user] = userdict[user][item]
print(itemdict)

itemavg = {}
for item in itemdict:
    sum = 0
    for user in itemdict[item]:
        sum += itemdict[item][user]
    itemavg[item] = sum / len(itemdict[item])
print(itemavg)

itemdict2 = {}
for item in itemdict:
    thislist = []
    for i in range(1, 13):
        user = 'user' + str(i)
        if user in itemdict[item]:
            thislist.append(itemdict[item][user] - itemavg[item])
        else:
            thislist.append(0)
    itemdict2[item] = thislist
print(itemdict2)

# 이 itemdict2를 가지고 row들간에 cosine similarities 계산
for i in range(1, 7):
    result = 1 - spatial.distance.cosine(itemdict2['item1'], itemdict2['item'+str(i)])
    print ("Sim(1,",str(i),") = ", result)
```
