```python
import networkx as nx
g = nx.path_graph(20)
```

```python
import matplotlib.pyplot as plt

nx.draw(g)
plt.show()
```

```python
pos = nx.spring_layout(g) # 펴서 그려라?

nx.draw_networkx_nodes(g, pos, node_color='grey', node_size=500, alpha=0.8) # 노드를 그려라 (g, pos는 필수고 나머진 옵션)
nx.draw_networkx_edges(g, pos, width=1.0, alpha=0.1) # 엣지를 그려라

labels = {}
labels[0] = '0'
labels[1] = '1'
nx.draw_networkx_labels(g, pos, labels, font_size=16) # 글자를 써라

plt.show()
```
